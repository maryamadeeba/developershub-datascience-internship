# Task 1: Exploring and Visualizing the Iris Dataset
## Introduction
This notebook explores the classic Iris dataset to understand data structures, relationships, and distributions using pandas and visualizations with matplotlib & seaborn.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Styling
sns.set_theme(style='whitegrid', palette='Set2')
plt.rcParams['figure.figsize'] = (8, 5)
```

## 1. Load Dataset

```python
iris = load_iris(as_frame=True)
df = iris.frame
df['species'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
df.drop(columns='target', inplace=True)
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
df.head()
```

## 2. Dataset Structure & Summary

```python
print("=== Data Types ===")
print(df.dtypes)
print("\n=== Statistical Summary ===")
df.describe()
```

```python
print("Missing Values:")
print(df.isnull().sum())
print("\nClass Distribution:")
print(df['species'].value_counts())
```

## 3. Exploratory Data Analysis (EDA)
### 3.1 Scatter Plot — Sepal Length vs Sepal Width

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
for ax, (x, y) in zip(axes, [('sepal length (cm)', 'sepal width (cm)'),
                               ('petal length (cm)', 'petal width (cm)')]):
    for species, grp in df.groupby('species'):
        ax.scatter(grp[x], grp[y], label=species, alpha=0.75)
    ax.set_xlabel(x); ax.set_ylabel(y)
    ax.set_title(f'{x} vs {y}')
    ax.legend()
plt.tight_layout()
plt.savefig('task1_scatter.png', dpi=100, bbox_inches='tight')
plt.show()
print("Scatter plots saved.")
```

### 3.2 Histograms — Feature Distributions

```python
features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for ax, feat in zip(axes.flatten(), features):
    for species, grp in df.groupby('species'):
        ax.hist(grp[feat], bins=15, alpha=0.6, label=species)
    ax.set_title(f'Distribution of {feat}')
    ax.set_xlabel(feat); ax.set_ylabel('Frequency')
    ax.legend()
plt.suptitle('Feature Distributions by Species', fontsize=14, y=1.01)
plt.tight_layout()
plt.savefig('task1_histograms.png', dpi=100, bbox_inches='tight')
plt.show()
```

### 3.3 Box Plots — Outlier Detection

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for ax, feat in zip(axes.flatten(), features):
    df.boxplot(column=feat, by='species', ax=ax, grid=False)
    ax.set_title(feat); ax.set_xlabel('Species')
plt.suptitle('Box Plots by Species', fontsize=14)
plt.tight_layout()
plt.savefig('task1_boxplots.png', dpi=100, bbox_inches='tight')
plt.show()
```

### 3.4 Correlation Heatmap

```python
plt.figure(figsize=(7, 5))
corr = df.drop(columns='species').corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.savefig('task1_heatmap.png', dpi=100, bbox_inches='tight')
plt.show()
```

### 3.5 Pairplot

```python
pair = sns.pairplot(df, hue='species', diag_kind='kde')
pair.fig.suptitle('Pairplot of Iris Features', y=1.02)
plt.savefig('task1_pairplot.png', dpi=100, bbox_inches='tight')
plt.show()
```

## 4. Conclusion

- **Petal length** and **petal width** are the most discriminative features for separating species.
- **Setosa** is clearly separable from the other two species in almost all feature combinations.
- Versicolor and Virginica show some overlap, especially in sepal dimensions.
- No missing values were found; the dataset is clean and well-balanced with 50 samples per class.
- The petal features have the highest inter-class correlation, making them the most useful for classification tasks.
